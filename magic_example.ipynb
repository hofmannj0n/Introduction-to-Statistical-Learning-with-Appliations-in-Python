```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

cols = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConcl', 'fAsym', 'fM3long', 'fM3Trans', 'fAlpha', 'fDist', 'class']

df = pd.read_csv('magic04.data', names=cols)

df['class'] = (df['class'] == 'g').astype(int)

df
```

```python
for label in cols[:-1]:
  plt.hist(df[df['class']==1][label], color='blue', density=True, label='gamma', alpha=0.7)
  plt.hist(df[df['class']==0][label], color='red', density=True, label='hadron', alpha=0.7)
  plt.title(label)
  plt.ylabel('probability')
  plt.xlabel(label)
  plt.legend
  plt.show()
```
